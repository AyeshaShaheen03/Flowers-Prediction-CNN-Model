```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
```

```python
TRAIN_DIR = "D:/AI projects/Neural-Network/cnn/seg_train/seg_train"
TEST_DIR = "D:/AI projects/Neural-Network/cnn/seg_test/seg_test"

# for cnn, it use for train
# Training generator
train_gen = ImageDataGenerator(rescale=1./255).flow_from_directory(
    TRAIN_DIR,
    target_size=(200, 200),
    batch_size=32,
    class_mode='categorical'
)

# Testing generator
test_gen = ImageDataGenerator(rescale=1./255).flow_from_directory(
    TEST_DIR,
    target_size=(200, 200),
    batch_size=32,
    class_mode='categorical'
)
```

```python
# for extra information, not required for model training
num_classes = len(train_gen.class_indices)
class_names = list(train_gen.class_indices.keys())
print("Class names:" , class_names)
print("Number of classes:" , num_classes )
print("Classes_directory:" , train_gen.class_indices)
```

```python
model = Sequential([
    # 32 means, 32 diff kernals houn gy with 3 by 3 metrix
    # relu is for which info or which info will precessand will pass to next, neg value to 0 mai convert
    # input shape means img size 200, 200, 3, 3 means gb  colors
    Conv2D(32, (3, 3), activation='relu', input_shape=(200, 200, 3)),
      MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(2, 2), Flatten(),
    Dense(128, activation='relu'),
    # num_classes means output neurons, you can also write 6 here
    Dense(num_classes, activation='softmax')
])
```

```python
model.summary()
```

```python
# train model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#history is vaiable here
history = model.fit(train_gen, epochs=30, validation_data=test_gen, batch_size = 32)
```

```python
loss, acc = model.evaluate(test_gen)
print("Test Loss:" , loss)
print("Test Accuracy:" , acc)
```

```python
model.save("model.h5")
```
